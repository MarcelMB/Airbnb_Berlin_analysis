### host_response_time + host_response_rate: deal with NaNs (have same amount of Nans)

```python
listings_filtered.host_response_time.value_counts(dropna=False)
```

```python
listings_filtered.host_response_rate.value_counts(dropna=False)
```

```python
#transfer to numerical
listings_filtered.host_response_rate=listings_filtered.host_response_rate.str.rstrip('%').astype('float')
```

```python
# a Nan value in reponse rate and time can occur if the client never reached out to the host, or the host never reponsed
#host reponse rate are sometimes zero, so nans are not that the host didnt respond
#this means the client never reached out
```

```python
#host_response_time Nans= customer never asked something
```

```python
listings_filtered.host_response_time=listings_filtered.host_response_time.fillna('no request')
listings_filtered.host_response_time.value_counts(dropna=False)
```

```python
#fill Nans with -1 for binning
listings_filtered.host_response_rate=listings_filtered.host_response_rate.fillna(-1)
```

```python
listings_filtered.host_response_rate
```

```python
# bin host_response_rate
rate_labels = ['No request','Low', 'Moderate', 'High']
listings_filtered['host_response_rate_category']=pd.cut(listings_filtered.host_response_rate,[-2,-1,50,80,100],labels=rate_labels)
#data[['IC2','ic2_category_userdef']].head(20)
```

```python
listings_filtered.host_response_rate_category.value_counts(dropna=False)
```

```python
#now all nans are indicated as no request and others categorized in different response rates
```

```python
#drop original reespone rate and time
listings_filtered.dropna(subset = ["host_response_rate"], inplace=True)
listings_filtered.dropna(subset = ["host_response_time"], inplace=True)
```

### bathrooms_text: deal with NaNs, create new columns bathroom type and quantity

```python
listings_filtered.bathrooms_text.value_counts(dropna=False)
```

```python
#we only have 8 Nans(drop those)
listings_filtered.dropna(subset = ["bathrooms_text"], inplace=True)
```

```python
# split this into two seperate columns: type of bath room (private, shared) & number of bathrooms (float)
```

```python
#0 shared baths add to 1 private bath
listings_filtered["bathrooms_text"] = listings_filtered["bathrooms_text"].str.replace('0 shared baths','1 private bath')
#0 baths add to 1 shared bath
listings_filtered["bathrooms_text"] = listings_filtered["bathrooms_text"].str.replace('0 baths','1 shared bath')
```

```python
#add half-bath to 1 shared bath (because whats half a bath, could mean shared)
listings_filtered["bathrooms_text"] = listings_filtered["bathrooms_text"].str.replace('Half-bath','1 shared bath')
```

```python
#shared half-bath add to 1 shared bath
listings_filtered["bathrooms_text"] = listings_filtered["bathrooms_text"].str.replace('Shared half-bath','1 shared bath')
```

```python
#private half-bath add to 1 private
listings_filtered["bathrooms_text"] = listings_filtered["bathrooms_text"].str.replace('Private half-bath','1 private bath')
```

```python
#find out what most frequent categroy 1 bath means is it private or shared
listings_filtered['property_type'].loc[listings_filtered['bathrooms_text'].str.contains("1 bath", case=False)].head(20)
```

```python
#seems like entire place is rented therefore its a private bathroom
listings_filtered["bathrooms_text"] = listings_filtered["bathrooms_text"].str.replace('1 private bath','1 bath')
```

```python
listings_filtered.bathrooms_text.value_counts(dropna=False)
```

```python
#if second splitted string is 'shared' put it into category shared
#otherwise its private catgeory

def clean_bath(x):
    if x.split(" ")[1]=='shared':
        return 'shared'
    else:
        return 'private'
```

```python
listings_filtered['bathrooms_type']=list(map(clean_bath, listings_filtered['bathrooms_text']))
```

```python
#get bathroom quantity and put it in a column
listings_filtered["bathroom_qty"] = listings_filtered["bathrooms_text"].str.split(" ", expand=True)[0]
```

```python
listings_filtered[['bathrooms_text','bathrooms_type','bathroom_qty']].head(15)
```

```python
#drop original bathroom text column
listings_filtered.dropna(subset = ["bathrooms_text"], inplace=True)
```

### deal with several review score columns with NaNs 

```python
#review_scores_accuracy   review_scores_cleanliness    review_scores_checkin
#review_scores_communication  review_scores_location   review_scores_value
```

```python
# listings with zero reviews are already filtered out earlier
#the number of nans in all these columns is quite low
```

```python
# this would be the overall rating
listings_filtered.review_scores_rating.describe()
```

```python
#it could be that some who gave a review didnt fill out the detail ratings for cleaniness, check in and so on
#fill Nans in those detailed despction with the mean of the overall rating
```

```python
listings_filtered.review_scores_cleanliness.describe()
#ratings go from zero to 5
#ad here its very similar to the overall rating already
```

```python
for col in listings_filtered[['review_scores_accuracy','review_scores_cleanliness','review_scores_checkin','review_scores_communication','review_scores_location','review_scores_value']]:
    listings_filtered[col] = listings_filtered[col].fillna(listings_filtered.review_scores_rating.median())
```
